```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data
data_path="../data/processed/processed_loan_data.csv"
df=pd.read_csv(data_path)
```

```python
print(df.shape)
```

```python
#checking target distribution
plt.figure(figsize=(6,4))
sns.countplot(x="loan_status",data=df)
plt.title("Loan Status Distribution (0 = High Risk, 1 = Low Risk)")
plt.show()
```

```python
#histograms for key_numerical featurs
num_cols = ["person_age","person_income","loan_amnt","credit_score","loan_int_rate","debt_to_income"]
df[num_cols].hist(bins=30, figsize=(12,8))
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
#boxplots

plt.figure(figsize=(12,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2,3,i)
    sns.boxplot(x='loan_status', y=col, data=df)
    plt.title(f"{col} vs Loan Status")

plt.show()
```

```python
#correlational heatmap
plt.figure(figsize=(10,8))
corr=df[num_cols + ['loan_status']].corr()  
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlatinal Heatmap")
plt.show()
```

```python
# Class imbalance check
class_counts = df['loan_status'].value_counts()
print("\nClass counts:\n", class_counts)
print("\nClass imbalance ratio:", class_counts[0]/class_counts[1])
```
